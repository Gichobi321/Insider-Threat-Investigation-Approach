```python
import matplotlib.pyplot as plt
import numpy as np
```

data

```python

months = ['07/2019', '08/2019', '09/2019', '10/2019', '11/2019']
searches = [50, 53, 59, 56, 62]
direct = [39, 47, 42, 51, 51]
social_media = [70, 80, 90, 87, 92]

x = np.arange(len(months))
width = 0.25
```

```python

# Plot
plt.figure(figsize=(10,6))
plt.bar(x - width, searches, width, label='Searches', color='dodgerblue')
plt.bar(x, direct, width, label='Direct', color='palevioletred')
plt.bar(x + width, social_media, width, label='Social Media', color='orange')

# Labels
plt.xlabel('months')
plt.ylabel('visitors\nin thousands', rotation=0, labelpad=50)
plt.title('Visitors by web traffic sources')
plt.xticks(x, months)
plt.legend()
plt.tight_layout()

plt.show()
```
